```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
movies=pd.read_csv("moviesL.csv")

def repl(str1):
    return str1.replace('|',' ')    

movies.genres=movies.genres.apply(repl)

movierate=[np.nan]*len(movies)
movies=movies.assign(rating=movierate)

movies.set_index('movieId',inplace=True)
movies
```

```python
ratings=pd.read_csv("ratingsL.csv")
del ratings['timestamp']
avg_ratings=ratings.groupby('movieId').mean()
avg_ratings.reset_index(inplace = True) 
del avg_ratings['userId']
avg_ratings.set_index('movieId',inplace=True)
avg_ratings
```

```python
for i in avg_ratings.index:
    movies.rating[i]=avg_ratings.rating[i]

movies.dropna(how='any',inplace=True)
```

```python
movies
```

```python
genre_list=['Action','Adventure','Animation','Children','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']

ls2=[]
for x in genre_list:
    ls1=[]
    for i in movies.index:
        if x in movies.genres[i]:
            ls1.append(1)
        else:
            ls1.append(0)
    ls2.append(ls1)
    
for i in range(0,len(genre_list)):
    movies[genre_list[i]]=ls2[i]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaled_movies=scaler.fit_transform(movies[['rating', 'Action', 'Adventure',
       'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
       'Sci-Fi', 'Thriller', 'War', 'Western']])
```

```python
movies.index
```

```python
scaled_movies=pd.DataFrame(scaled_movies,columns=['rating', 'Action', 'Adventure',
       'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
       'Sci-Fi', 'Thriller', 'War', 'Western'],index=movies.index)
```

```python
scaled_movies
```

```python
inertias=[]
for k in range(1,101):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit_transform(scaled_movies[['rating', 'Action', 'Adventure',
       'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
       'Sci-Fi', 'Thriller', 'War', 'Western']])
    inertias.append(kmeans.inertia_)
```

```python
plt.scatter(range(1,101),inertias)
plt.show()
```

```python
title_list=list(movies.title)
input_movies_id=[]

for i in range(0,10):
    mov=input_movies[i]
    for j in movies.index:
        if mov==movies.title[j]:
            input_movies_id.append(j)
```

```python
input_movies_id
```

```python
cluster_list=[]

for i in range(0,10):
    cluster_list.append(movies.cluster[input_movies_id[i]])
```

```python
cluster_list=set(cluster_list)
cluster_list=list(cluster_list)
cluster_list
```

```python
indices=list(movies.index[::-1])
```

```python
recommend=[]

for i in cluster_list:
    for j in indices:
        if movies.cluster[j]==i:
            recommend.append(movies.title[j])
            break
```

import pandas as pd 
import numpy as np
movies=pd.read_csv("movies.csv")

def repl(str1):
    return str1.replace('|',' ')    

movies.genres=movies.genres.apply(repl)

from sklearn.feature_extraction.text import TfidfVectorizer
Vec=TfidfVectorizer()
genres=Vec.fit_transform(movies.genres)

from sklearn.metrics.pairwise import cosine_similarity
co_sim = cosine_similarity(genres, genres)
co_sim

```python
movies['ind']=list(movies.index)

# name=input("Enter the name of the movie ")
x=movies[['ind']].loc[movies['title']=='Toy Story (1995)']
y=(x.index).tolist()[0]

ls2=[]
ls1=list(co_sim[y])
for i in range(len(ls1)):
    if(ls1[i]>=0.95):
        ls2.append(i)

movies.loc[ls2]
```
